# 2. DATA PREPROCESSING

## 2.1 Initial Cleaning Data

```python
# --- XỬ LÝ FILE MDL ---

# Xử lý Title phim
df_mdl['Title'] = df_mdl['Title'].str.lower().str.strip().str.replace(r'[^\w\s]', '', regex=True)

# Tách cột Aired thành
def extract_dates(text):
    if pd.isna(text) or text == 'N/A':
        return [None] * 4

    # Tách chuỗi bằng dấu gạch ngang " - "
    parts = str(text).split(' - ')

    def get_info(date_str):
        # Tìm tháng (3 chữ cái) và năm (4 chữ số)
        m = re.search(r'([A-Z][a-z]{2})', date_str)
        y = re.search(r'(\d{4})', date_str)
        return (m.group(1) if m else None, int(y.group(1)) if y else None)

    # Lấy thông tin ngày bắt đầu
    s_month, s_year = get_info(parts[0])

    # Xử lý ngày kết thúc
    if len(parts) > 1:
        e_month, e_year = get_info(parts[1])
    else:
        # Nếu phim chỉ chiếu trong 1 ngày (ví dụ Movie), kết thúc = bắt đầu
        e_month, e_year = s_month, s_year

    return s_month, s_year, e_month, e_year

# Tạo các cột mới
df_mdl[['start_month', 'start_year', 'end_month', 'end_year']] = df_mdl['Aired'].apply(
    lambda x: pd.Series(extract_dates(x))
)

# Chuyển Duration về số phút
def convert_duration(duration):
    if pd.isna(duration): return duration
    hours = re.search(r'(\d+)\s*hr', str(duration))
    mins = re.search(r'(\d+)\s*min', str(duration))
    total = 0
    if hours: total += int(hours.group(1)) * 60
    if mins: total += int(mins.group(1))
    return total if total > 0 else None

df_mdl['duration_mins'] = df_mdl['Duration'].apply(convert_duration)

# Tách Score thành rating (float) và reviews_count (int)
def split_score_detailed(text):
    if pd.isna(text) or text == 'N/A':
        return 0.0, 0
    # Lấy số thực đầu tiên làm rating
    rating_match = re.search(r'(\d+\.\d+)', str(text))
    # Lấy dãy số có dấu phẩy nằm sau "by " làm số lượng reviews
    reviews_match = re.search(r'by ([\d,]+)', str(text))

    rating = float(rating_match.group(1)) if rating_match else 0.0
    reviews = int(reviews_match.group(1).replace(',', '')) if reviews_match else 0
    return rating, reviews

df_mdl[['rating', 'reviews_count']] = df_mdl['Score'].apply(
    lambda x: pd.Series(split_score_detailed(x))
)

# Chuyển Ranked và Popularity thành kiểu số nguyên (Integer)
def clean_to_int(text):
    if pd.isna(text):
        return None
    # Xóa ký tự '#' và dấu phẩy, chỉ giữ lại số
    val = re.sub(r'[^\d]', '', str(text))
    return int(val) if val != '' else None

df_mdl['Ranked'] = df_mdl['Ranked'].apply(clean_to_int)
df_mdl['Popularity'] = df_mdl['Popularity'].apply(clean_to_int)

# Tách Cast
# n=2 nghĩa là chỉ tách tối đa 2 lần để lấy 3 phần (Lead1, Lead2, và phần còn lại)
cast_split = df_mdl['Cast'].str.split(',', n=2, expand=True)
df_mdl['main_lead1'] = df_mdl['Cast'].str.split(',').str[0].str.strip()
df_mdl['main_lead2'] = df_mdl['Cast'].str.split(',').str[1].str.strip()
df_mdl['other_cast'] = cast_split[2].str.strip()

# Làm sạch khoảng trắng thừa
df_mdl['main_lead1'] = df_mdl['main_lead1'].str.strip()
df_mdl['main_lead2'] = df_mdl['main_lead2'].str.strip()

# Tính tổng số lượng diễn viên (Tính từ cột Cast gốc)
df_mdl['cast_count'] = df_mdl['Cast'].fillna('').apply(
    lambda x: len(x.split(',')) if x != '' else 0
)

# Kiểm tra xem có bao nhiêu dòng bị trống ở cột mới
print(df_mdl[['main_lead1', 'main_lead2']].isnull().sum())

# Tạo cột platform_count bằng cách đếm số lượng mục trong Watch Platforms
df_mdl['platform_count'] = df_mdl['Watch Platforms'].fillna('').apply(
    lambda x: len([p for p in x.split(',') if p.strip() != '']) if x != '' else 0
)

# Tách lấy phần trước dấu gạch ngang và xóa khoảng trắng thừa
df_mdl['Content Rating'] = df_mdl['Content Rating'].fillna('').apply(
    lambda x: x.split('-')[0].strip() if '-' in x else x
)
df_mdl['Content Rating'] = df_mdl['Content Rating'].str.replace(r'18\+.*', '18+', regex=True)

# Xóa các cột không cần thiết
columns_to_drop = ['Cast', 'Watch Platforms', 'Aired', 'Aired On', 'URL', 'other_cast', 'Duration', 'Score']
df_mdl.drop(columns=columns_to_drop, inplace=True, errors='ignore')

df_mdl = df_mdl.rename(columns={
    'Title': 'title', 'Synopsis': 'content', 'Screenwriters': 'screenwriters',
    'Directors': 'directors', 'Genres': 'genres', 'Tags': 'tags', 'Episodes': 'episodes',
    'Watchers': 'watchers', 'Original Network': 'original_network', 'Content Rating': 'age_rating',
    'Ranked': 'ranked', 'Popularity': 'popularity'
})

# Xem cấu trúc bảng cuối cùng
print("\nCác cột còn lại sau khi tối ưu:")
print(df_mdl.columns.tolist())
display(df_mdl.head(3))
```

```python
# --- XỬ LÝ FILE KGR (KDRAMAS.CSV) ---

# Xử lý Title phim
df_kg['title'] = df_kg['title'].str.lower().str.strip()

# Sửa lỗi thiếu khoảng trắng trong cột genres
df_kg['genres'] = df_kg['genres'].astype(str).str.replace(r',\s*', ', ', regex=True)

# Chuyển cột votes từ string sang int
def convert_votes(vote_str):
    if pd.isna(vote_str): return 0
    vote_str = str(vote_str).upper().replace(',', '')
    if 'K' in vote_str:
        return int(float(vote_str.replace('K', '')) * 1000)
    elif 'M' in vote_str:
        return int(float(vote_str.replace('M', '')) * 1000000)
    return int(float(vote_str))

df_kg['votes'] = df_kg['votes'].apply(convert_votes)

# Hàm dùng chung để làm sạch danh sách và định dạng tên người
def clean_person_name(x):
    if pd.isna(x) or x == "['\\\\N']" or x == "[]" or x == "\\N":
        return np.nan
    try:
        # Nếu là dạng chuỗi của list "['A', 'B']", chuyển về list thật
        if isinstance(x, str) and x.startswith('['):
            actual_list = ast.literal_eval(x)
            x = ", ".join(actual_list)

        # 1. Xóa dấu gạch ngang và khoảng trắng thừa
        cleaned = str(x).replace('-', ' ').strip()

        # 2. Xử lý viết hoa:
        # Nếu là 'iu' (bất kể hoa thường) thì giữ nguyên 'IU'
        if cleaned.lower() == 'iu':
            return 'IU'

        # Ngược lại thì mới dùng .title() cho các tên thông thường
        return cleaned.title()
    except:
        # Tương tự cho phần ngoại lệ
        cleaned = str(x).replace('-', ' ').strip()
        return 'IU' if cleaned.lower() == 'iu' else cleaned.title()

# 2. Áp dụng làm sạch cho các cột tên người
person_cols = ['writer', 'director', 'mainLead1', 'mainLead2']
for col in person_cols:
    df_kg[col] = df_kg[col].apply(clean_person_name)

# 3. Xử lý logic trùng lặp diễn viên chính trong cùng 1 bộ phim ở 2 cột mainLead1 và mainLead2
df_kg.loc[df_kg['mainLead1'] == df_kg['mainLead2'], 'mainLead2'] = np.nan

# Xử lý các kí tự \N trong startYear và endYear
# Thay thế '\N' thành NaN và chuyển về dạng số (float/int)
df_kg['startYear'] = pd.to_numeric(df_kg['startYear'].replace(r'\\N', np.nan, regex=True), errors='coerce')
df_kg['endYear'] = pd.to_numeric(df_kg['endYear'].replace(r'\\N', np.nan, regex=True), errors='coerce')

# Xử lý cột runtimeMinutes
# Chuyển '\N' thành NaN và ép kiểu về số
df_kg['runtimeMinutes'] = pd.to_numeric(df_kg['runtimeMinutes'].replace(r'\\N', np.nan, regex=True), errors='coerce')

# Xử lý cột age_rating
# Tạo từ điển quy đổi
age_map = {
    'TV-MA': '18+',
    'TV-14': '15+',
    'TV-PG': '12+',
    'TV-G': 'G',
    'TV-Y': 'G',
    'TV-Y7': '7+',
    'TV-Y7-FV': '7+',
    '16+': '15+',
    '13+': '13+',
    'MA-13': '13+',
    'Not Rated': np.nan
}

# Áp dụng quy đổi
df_kg['age_rating'] = df_kg['age_rating'].map(age_map)

# Xóa các cột không cần thiết
columns_to_drop = ['id', 'image_url']
df_kg.drop(columns=columns_to_drop, inplace=True, errors='ignore')

df_kg = df_kg.rename(columns={
    'synopsis': 'content', 'startYear': 'start_year',
    'endYear': 'end_year', 'writer': 'screenwriters', 'mainLead1': 'main_lead1',
    'mainLead2': 'main_lead2', 'director': 'directors', 'runtimeMinutes': 'duration_mins'
})

# Xem cấu trúc bảng cuối cùng
print("\nCác cột còn lại sau khi tối ưu:")
print(df_kg.columns.tolist())
display(df_kg.head(3))
```

## 2.2 Data Merging

```python
# --- GỘP VÀ LẤP ĐẦY DỮ LIỆU (CHỐNG MẤT DỮ LIỆU) ---

# Hàm dọn dẹp tiêu đề chuyên sâu (xóa ký tự đặc biệt, "season", khoảng trắng)
def super_clean_title(text):
    if pd.isna(text): return ""
    text = str(text).lower()
    text = text.replace('season', '').replace('special', '')
    # Chỉ giữ lại chữ cái và số
    text = re.sub(r'[^a-z0-9]', '', text)
    return text.strip()

# Chuẩn bị bản tạm
df_mdl_temp = df_mdl.copy()
df_kg_temp = df_kg.copy()

# Sử dụng hàm super_clean_title đã định nghĩa ở bước trước
df_mdl_temp['title_key'] = df_mdl_temp['title'].apply(super_clean_title)
df_kg_temp['title_key'] = df_kg_temp['title'].apply(super_clean_title)

# Gộp bảng KG vào bảng MDL dựa trên Tên và Năm đã làm sạch
df_combined = pd.merge(
    df_mdl_temp,
    df_kg_temp,
    on=['title_key'],
    how='left',
    suffixes=('_mdl', '_kg')
)

display(df_combined.head(1))

# Tính số lượng phim tìm thấy thông tin từ bảng KG
matched = df_combined['title_kg'].notna().sum()
total = len(df_combined)
print(f"Tỷ lệ khớp thành công: {matched}/{total} ({matched/total*100:.2f}%)")
```

## 2.3 Post-Merge Data Cleaning

```python
# VÁ THÔNG TIN CHO BẢNG MDL

# cột content (Nội dung phim) - MDL thiếu 123
df_combined['content_mdl'] = df_combined['content_mdl'].fillna(df_combined['content_kg'])

# cột screenwriters - MDL thiếu 975
df_combined['screenwriters_mdl'] = df_combined['screenwriters_mdl'].fillna(df_combined['screenwriters_kg'])

# cột directors - MDL thiếu 700
df_combined['directors_mdl'] = df_combined['directors_mdl'].fillna(df_combined['directors_kg'])

# cột genres - MDL thiếu 5
df_combined['genres_mdl'] = df_combined['genres_mdl'].fillna(df_combined['genres_kg'])

# cột main_lead1 và main_lead2 - MDL thiếu 108
df_combined['main_lead1_mdl'] = df_combined['main_lead1_mdl'].fillna(df_combined['main_lead1_kg'])
df_combined['main_lead2_mdl'] = df_combined['main_lead2_mdl'].fillna(df_combined['main_lead2_kg'])

# cột duration_mins - MDL thiếu 33
df_combined['duration_mins_mdl'] = df_combined['duration_mins_mdl'].fillna(df_combined['duration_mins_kg'])

# cột start_year và end_year - MDL thiếu 304
df_combined['start_year_mdl'] = df_combined['start_year_mdl'].fillna(df_combined['start_year_kg'])
df_combined['end_year_mdl'] = df_combined['end_year_mdl'].fillna(df_combined['end_year_kg'])

# Kiểm tra missing value sau khi vá
print('Check for missing value - df_mdl')
display(df_mdl[['content', 'screenwriters', 'directors', 'genres', 'main_lead1', 'main_lead2', 'duration_mins', 'platform_count', 'original_network', 'start_month', 'start_year', 'end_month', 'end_year']].isnull().sum())

print('Check for missing value - df_combined')
display(df_combined[['content_mdl', 'screenwriters_mdl', 'directors_mdl', 'genres_mdl', 'main_lead1_mdl', 'main_lead2_mdl', 'duration_mins_mdl', 'platform_count', 'original_network', 'start_month', 'start_year_mdl', 'end_month', 'end_year_mdl', 'votes']].isnull().sum())

# Giữ lại các cột _mdl và xóa các cột _kg dư thừa
cols_to_keep = [c for c in df_combined.columns if not c.endswith('_kg') and c not in ['title_key', 'start_year_key']]
df_final = df_combined[cols_to_keep].copy()

# Đổi tên (bỏ đuôi _mdl)
df_final.columns = [c.replace('_mdl', '') for c in df_final.columns]



print("\nCác cột còn lại sau khi tối ưu:")
print(df_final.columns.tolist())
display(df_final.head(1))
```

```python
# XỬ LÝ CÁC Ô TRỐNG

# Danh sách các cột dạng chữ cần điền "Unknown"
text_cols = [
    'content', 'screenwriters', 'directors',
    'genres', 'main_lead1', 'main_lead2',
    'original_network'
]

df_final[text_cols] = df_final[text_cols].fillna('Unknown')

# Điền giá trị trung vị (median) cho thời lượng phim
df_final['duration_mins'] = df_final['duration_mins'].fillna(df_final['duration_mins'].median())

# Điền 0 cho những phim không có thông tin lượt bình chọn
df_final['votes'] = df_final['votes'].fillna(0)

# Vá end_year bằng start_year (Logic thực tế)
df_final['end_year'] = df_final['end_year'].fillna(df_final['start_year'])

# Vá các giá trị số còn lại bằng giá trị phổ biến nhất (Mode)
df_final['start_month'] = df_final['start_month'].fillna(df_final['start_month'].mode()[0])
df_final['end_month'] = df_final['end_month'].fillna(df_final['end_month'].mode()[0])

# Kiểm tra missing value
print('Check for missing value - df_combined')
display(df_final.isnull().sum())
```

# 3.1 Multi-label Encoding




```python
from sklearn.preprocessing import MultiLabelBinarizer, LabelEncoder
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.model_selection import train_test_split

# Định nghĩa hàm này ở phía trên cùng của code
def my_tokenizer(text):
    return text.split()

# 1. Load dữ liệu
df = pd.read_csv('kdrama_final_cleaned.csv')

# --- TIỀN XỬ LÝ CƠ BẢN (TRƯỚC KHI CHIA TẬP) ---

# Xử lý cột "watchers": Chuyển từ chuỗi "81,971" sang số thực
if df['watchers'].dtype == 'object':
    df['watchers'] = df['watchers'].str.replace(',', '').astype(float)

# Log Transform cho các biến có giá trị lớn để giảm độ lệch (Skewness)
df['watchers_log'] = np.log1p(df['watchers'])
df['popularity_log'] = np.log1p(df['popularity'])

# Xử lý thời gian (Cyclical Encoding cho tháng)
df['start_month_sin'] = np.sin(2 * np.pi * df['start_month'] / 12)
df['start_month_cos'] = np.cos(2 * np.pi * df['start_month'] / 12)
df['movie_age'] = 2026 - df['start_year']

# Xử lý Age Rating (Ordinal Mapping)
age_map = {'G': 1, '13+': 2, '15+': 3, '18+': 4, 'Unknown': 0, 'Not Yet Rated': 0}
df['age_rating_val'] = df['age_rating'].map(age_map).fillna(0)

# --- CHIA TẬP TRAIN VÀ TEST (BƯỚC QUAN TRỌNG NHẤT) ---
# Chúng ta chia sớm để các bước tính trung bình (Target Encoding) không bị "nhìn trước" tập Test
df_train, df_test = train_test_split(df, test_size=0.2, random_state=42)

# Tạo bản sao để tránh cảnh báo SettingWithCopyWarning
df_train = df_train.copy()
df_test = df_test.copy()

# --- TARGET ENCODING (CHỈ TÍNH TRÊN TRAIN, APPLY CHO TEST) ---

def apply_target_encoding(train_df, test_df, column, target):
    # Tính trung bình target của cột đó trên tập TRAIN
    target_mean = train_df.groupby(column)[target].mean()
    global_mean = train_df[target].mean() # Giá trị trung bình chung để điền cho người mới

    # Map vào tập Train
    train_df[f'{column}_score'] = train_df[column].map(target_mean)

    # Map vào tập Test (Những ai không có trong Train sẽ nhận global_mean)
    test_df[f'{column}_score'] = test_df[column].map(target_mean).fillna(global_mean)
    return train_df, test_df

# Áp dụng cho Diễn viên, Đạo diễn, Biên kịch
# Lưu ý: Nếu bạn dự đoán Rating, không nên dùng Rating làm Target ở đây.
# Ở đây tôi dùng 'rating' làm cơ sở tính score để dự đoán 'popularity' và 'watchers'.
target_to_use = 'rating'
for col in ['main_lead1', 'main_lead2', 'directors', 'screenwriters']:
    df_train, df_test = apply_target_encoding(df_train, df_test, col, target_to_use)

# --- XỬ LÝ GENRES, TAGS, CONTENT (TF-IDF & MLB) ---

# Genres (MultiLabelBinarizer)
mlb = MultiLabelBinarizer()
df_train['genres_list'] = df_train['genres'].fillna('').apply(lambda x: [i.strip() for i in x.split(',')])
df_test['genres_list'] = df_test['genres'].fillna('').apply(lambda x: [i.strip() for i in x.split(',')])

train_genres = mlb.fit_transform(df_train['genres_list'])
test_genres = mlb.transform(df_test['genres_list']) # Chỉ transform tập test

# Tags (TF-IDF)
# tfidf_tag = TfidfVectorizer(max_features=100, tokenizer=lambda x: x.split(), token_pattern=None)
tfidf_tag = TfidfVectorizer(max_features=100, tokenizer=my_tokenizer, token_pattern=None)
# Xóa nhiễu (Vote tags)
def clean_tags(text):
    return str(text).replace('(Vote tags)', '').replace(',', ' ')

train_tags = tfidf_tag.fit_transform(df_train['tags'].apply(clean_tags))
test_tags = tfidf_tag.transform(df_test['tags'].apply(clean_tags))

# Content (TF-IDF)
tfidf_content = TfidfVectorizer(max_features=100, stop_words='english')
train_content = tfidf_content.fit_transform(df_train['content'].fillna(''))
test_content = tfidf_content.transform(df_test['content'].fillna(''))

# --- KẾT HỢP TẤT CẢ LẠI THÀNH FILE CUỐI CÙNG ---

def finalize_df(df_part, genres_arr, tags_arr, content_arr, mlb_cols, tags_cols, content_cols):
    # Lấy các cột số đã xử lý
    numeric_features = [
        'episodes', 'duration_mins', 'start_year', 'movie_age', 'age_rating_val',
        'main_lead1_score', 'main_lead2_score', 'directors_score', 'screenwriters_score',
        'start_month_sin', 'start_month_cos', 'watchers_log', 'popularity_log', 'rating'
    ]

    res = df_part[numeric_features].reset_index(drop=True)

    # Thêm các cột từ Sparse Matrix
    genres_df = pd.DataFrame(genres_arr, columns=[f"genre_{c}" for c in mlb_cols])
    tags_df = pd.DataFrame(tags_arr.toarray(), columns=[f"tag_{c}" for c in tags_cols])
    content_df = pd.DataFrame(content_arr.toarray(), columns=[f"txt_{c}" for c in content_cols])

    return pd.concat([res, genres_df, tags_df, content_df], axis=1)

X_train = finalize_df(df_train, train_genres, train_tags, train_content,
                     mlb.classes_, tfidf_tag.get_feature_names_out(), tfidf_content.get_feature_names_out())

X_test = finalize_df(df_test, test_genres, test_tags, test_content,
                    mlb.classes_, tfidf_tag.get_feature_names_out(), tfidf_content.get_feature_names_out())

print(f"Kích thước tập Train: {X_train.shape}")
print(f"Kích thước tập Test: {X_test.shape}")

display(X_train.head(5))
display(X_train.info())

# Danh sách các cột mục tiêu
target_cols = ['rating', 'watchers_log', 'popularity_log']

# Tạo y_train và y_test từ X_train và X_test hiện tại
y_train = X_train[target_cols].copy()
y_test = X_test[target_cols].copy()
```

```python
# --- XUẤT DỮ LIỆU PHỤC VỤ AI (METADATA) ---

# 1. Gộp tất cả diễn viên (Lead 1 và Lead 2) để tính điểm năng lực chung
all_leads = pd.concat([
    df_train[['main_lead1', 'rating']].rename(columns={'main_lead1': 'actor'}),
    df_train[['main_lead2', 'rating']].rename(columns={'main_lead2': 'actor'})
])
actor_scores = all_leads.groupby('actor')['rating'].mean().reset_index()
actor_scores.to_csv('actor_scores.csv', index=False)

# 2. Xuất điểm Đạo diễn
director_scores = df_train.groupby('directors')['rating'].mean().reset_index()
director_scores.to_csv('director_scores.csv', index=False)

# 3. Xuất điểm Biên kịch
writer_scores = df_train.groupby('screenwriters')['rating'].mean().reset_index()
writer_scores.to_csv('writer_scores.csv', index=False)

# 4. Lưu giá trị trung bình toàn cầu (Global Mean)
# Cái này cực quan trọng: Dùng khi người dùng nhập diễn viên mới tinh chưa có trong data
global_mean = df_train['rating'].mean()
with open('global_metadata.txt', 'w') as f:
    f.write(str(global_mean))

print("Đã xuất các file Metadata thành công!")
```

#3.2 Compute Correlation Matrix

```python
# Danh sách các cột "cấm" (chỉ có sau khi phim chiếu, không dùng để dự đoán)
forbidden_cols = ['rating', 'watchers', 'watchers_log', 'popularity', 'popularity_log',
                  'ranked', 'votes', 'reviews_count']

def get_best_features(X, target_series, name, threshold=0.05):
    # Chỉ tính tương quan với các cột KHÔNG nằm trong danh sách cấm
    potential_features = X.drop(columns=[c for c in forbidden_cols if c in X.columns], errors='ignore')

    # Tính tương quan
    correlations = potential_features.apply(lambda x: x.corr(target_series))

    # Lọc theo ngưỡng
    best_corr = correlations[abs(correlations) > threshold].sort_values(ascending=False)

    print(f"\n" + "="*60)
    print(f" CHI TIẾT TƯƠNG QUAN CHO: {name} (Ngưỡng > {threshold})")
    print(f" Tìm thấy {len(best_corr)} đặc trưng phù hợp.")
    print("="*60)
    print(best_corr)

    return best_corr.index.tolist()

# --- Thực hiện lọc cho từng Target ---

# 1. Tập đặc trưng cho Rating
features_rating = get_best_features(X_train, X_train['rating'], "RATING")
X_train_rating = X_train[features_rating]
X_test_rating = X_test[features_rating]

# 2. Tập đặc trưng cho Watchers
features_watchers = get_best_features(X_train, X_train['watchers_log'], "WATCHERS")
X_train_watchers = X_train[features_watchers]
X_test_watchers = X_test[features_watchers]

# 3. Tập đặc trưng cho Popularity
features_pop = get_best_features(X_train, X_train['popularity_log'], "POPULARITY")
X_train_pop = X_train[features_pop]
X_test_pop = X_test[features_pop]
```

# 4. TRAIN MODEL and EVALUATING

Sử dụng phiên bản **Ridge Regressio**n của  **Linear Regression**. Lý do là vì dù đã lọc xuống còn ~150 cột, nhưng số lượng đặc trưng vẫn còn khá nhiều so với số lượng dòng, Ridge sẽ giúp kiểm soát các hệ số để tránh mô hình bị "quá tải" (overfitting) và cho kết quả ổn định hơn trên website.:

```python
from sklearn.linear_model import Ridge
from sklearn.metrics import mean_absolute_error, r2_score, mean_squared_error
import numpy as np
import pandas as pd

# Lưu trữ kết quả để so sánh
performance_results = []
trained_models = {}

# Cấu hình các mục tiêu và tập dữ liệu tương ứng
tasks = [
    ("Rating", X_train_rating, X_test_rating, 'rating'),
    ("Watchers (Log)", X_train_watchers, X_test_watchers, 'watchers_log'),
    ("Popularity (Log)", X_train_pop, X_test_pop, 'popularity_log')
]

print("===== ĐANG HUẤN LUYỆN VÀ ĐÁNH GIÁ MÔ HÌNH RIDGE REGRESSION =====")

for name, X_tr, X_te, target_col in tasks:
    # 1. Khởi tạo mô hình Ridge (Linear Regression có bộ lọc)
    model = Ridge(alpha=1.0)

    # 2. Huấn luyện
    # Lưu ý: y_train và y_test phải lấy từ tập dữ liệu gốc ban đầu của bạn
    y_tr = y_train[target_col]
    y_te = y_test[target_col]

    model.fit(X_tr, y_tr)
    trained_models[name] = model

    # 3. Dự đoán trên tập Test
    y_pred = model.predict(X_te)

    # 4. Tính toán các chỉ số đo lường
    r2 = r2_score(y_te, y_pred)
    mae = mean_absolute_error(y_te, y_pred)
    rmse = np.sqrt(mean_squared_error(y_te, y_pred))

    performance_results.append({
        'Mô hình': name,
        'R2 Score': round(r2, 4),
        'MAE': round(mae, 4),
        'RMSE': round(rmse, 4)
    })

# In bảng kết quả tổng hợp
df_results = pd.DataFrame(performance_results)
print("\nBẢNG SO SÁNH ĐỘ CHÍNH XÁC:")
print(df_results.to_string(index=False))
print()

# Tính % độ chính xác dựa trên thang điểm 10 cho Rating
# Và dựa trên R2 Score cho các biến Log
print("===== ĐỘ CHÍNH XÁC CHI TIẾT (%) =====")

for result in performance_results:
    name = result['Mô hình']
    r2 = result['R2 Score']
    mae = result['MAE']

    if name == "Rating":
        # Với Rating, tính dựa trên sai số so với thang điểm 10
        acc_pct = (1 - (mae / 10)) * 100
        print(f"- {name}: {acc_pct:.2f}% (Dựa trên sai số điểm số)")
    else:
        # Với các biến Log, R2 Score phản ánh sát nhất độ chính xác của mô hình
        acc_pct = r2 * 100
        print(f"- {name}: {acc_pct:.2f}% (Độ khớp của mô hình - R2)")

print("-" * 40)
```

#5. HANDLING AFTER TRAINING

```python
import numpy as np

# Tính toán các mốc dựa trên tập dữ liệu gốc (df là dataframe ban đầu của bạn)
# Giả sử: Top 10% là Hot, 10-40% là Trung bình, còn lại là Thấp
threshold_hot = df['popularity'].quantile(0.1)     # Top 10% hạng đầu
threshold_medium = df['popularity'].quantile(0.4)  # Top 40% hạng đầu

print(f"Mốc để thành phim HOT: Hạng phải nhỏ hơn {int(threshold_hot)}")
print(f"Mốc để thành phim TRUNG BÌNH: Hạng nhỏ hơn {int(threshold_medium)}")

# --- HÀM CHUYỂN ĐỔI NHÃN ---
def get_popularity_status(rank_pred, threshold_hot, threshold_medium):
    """Chuyển đổi con số hạng dự đoán sang nhãn văn bản"""
    if rank_pred <= threshold_hot:
        return "🔥 HOT (Cực kỳ phổ biến)"
    elif rank_pred <= threshold_medium:
        return "📈 Trung bình"
    else:
        return "☁️ Thấp"

# --- HÀM DỰ ĐOÁN TỔNG HỢP ---
def predict_all_targets(models_dict, X_rating, X_watchers, X_pop, threshold_hot, threshold_medium):
    """Dự đoán cả 3 mục tiêu và trả về kết quả đã xử lý"""

    # 1. Dự đoán Rating (Giữ nguyên số thực)
    res_rating = models_dict['Rating'].predict(X_rating)[0]

    # 2. Dự đoán Watchers (Chuyển từ Log về số thực)
    log_watchers = models_dict['Watchers (Log)'].predict(X_watchers)[0]
    res_watchers = np.expm1(log_watchers)

    # 3. Dự đoán Popularity (Chuyển từ Log -> Hạng -> Nhãn)
    log_pop = models_dict['Popularity (Log)'].predict(X_pop)[0]
    rank_pop = np.expm1(log_pop)
    res_pop_label = get_popularity_status(rank_pop, threshold_hot, threshold_medium)

    return {
        'rating': round(float(res_rating), 2),
        'watchers': int(res_watchers),
        'popularity_level': res_pop_label,
    }
```

```python
import numpy as np

# Giả sử bạn lấy dữ liệu của bộ phim ở vị trí index 10 trong tập Test để thử
idx_to_test = X_test.index[10]

# 1. Lấy Index thực sự từ vị trí số 10
idx_to_test = X_test.index[10]

# Trích xuất dữ liệu đầu vào cho từng model (đúng các cột đã lọc)
input_r = X_test_rating.loc[[idx_to_test]]
input_w = X_test_watchers.loc[[idx_to_test]]
input_p = X_test_pop.loc[[idx_to_test]]

# GỌI DỰ ĐOÁN
prediction = predict_all_targets(
    trained_models,
    input_r, input_w, input_p,
    threshold_hot, threshold_medium
)

# IN KẾT QUẢ RA MÀN HÌNH
print(f"--- KẾT QUẢ DỰ ĐOÁN CHO PHIM (Index {idx_to_test}) ---")
print(f"⭐ Rating dự đoán: {prediction['rating']} / 10")
print(f"👥 Lượng người xem dự kiến: {prediction['watchers']:,} người")
print(f"📊 Mức độ phổ biến: {prediction['popularity_level']}")

# So sánh với thực tế (Nếu có)
print(f"\nThực tế: Rating {y_test.loc[idx_to_test, 'rating']} | Watchers {int(np.expm1(y_test.loc[idx_to_test, 'watchers_log'])):,}")
```
